Permitir a un usuario cargar los nombres de miembros de su familia permitiendo hasta 3 generaciones. Luego permitir seleccionar un miembro y una relación y devolver los familiares correspondiente. Ej "Pedro,hijo" debería imprimir los hijos de Pedro. Permitir relaciónes: padre, hijo, tio, hermano, abuelo, nieto.

```python
from anytree import Node, RenderTree

def return_is_root():
    is_parent_root = []

    for value in familia:
        if familia[value].is_root:
            is_parent_root.append(familia[value])
            
    return is_parent_root[0]
    
def renderizar_familia():            
    root = return_is_root()  
    print('\n')
    for row in RenderTree(root, childiter=reversed):
        print("%s%s" % (row.pre, row.node.name))

def buscar_padre(nodo):
    print(f'El padre de {nodo.name} es {nodo.parent.name}')

def buscar_hijos(nodo):
    print(f'Los hijos de {nodo.name} son:')
    for h in nodo.children:
        print(f'\t{h.name}')
        
def buscar_abuelo(nodo):
    padre = nodo.parent
    abuelo = padre.parent
    print(f'El abuelo de {nodo.name} es {abuelo.name}')
    
def buscar_tios(nodo):
    padre = nodo.parent
    tios = padre.siblings
    print(f'Los tios de {nodo.name} son:')
    for t in tios:
        print(f'\t{t.name}')

def buscar_hermanos(nodo):
    hermanos = nodo.siblings
    print(f'Los hermanos de {nodo.name} son:')
    for h in hermanos:
        print(f'\t{h.name}')
        
def buscar_nietos(nodo):
    hijos = nodo.children
    nietos = []
    for h in hijos:
        for n in h.children:
            nietos.append(n.name)
    
    print(f'Los nietos de {nodo.name} son:')
    for n in nietos:
        print(f'\t{n}')
        

if __name__ == '__main__':
	ejercicio = '0'
	familia = {}
	while ejercicio == '0':
		print('\n________________Menu________________')
		print('1: Agregar relación familiar\n2: Buscar relaciones\n3: Salir')
		accion = input('\nSeleccione la opcion deseada: ')
		if accion == '1':
			nombre = input('Nombre: ')
			if nombre in familia:
				print(f'el ya conocido {nombre}')
				node_p = familia[nombre]
				rol_familiar = input('Es un (hijo/padre): ')
				if rol_familiar == 'hijo':
					nombreP = input('De quien: ')
					if nombreP in familia:
						node_p.parent = familia[nombreP]
					else:
						familia[nombreP] = Node(nombreP)
						node_p.parent = familia[nombreP]
				elif rol_familiar == 'padre':
					nombreH = input('De quien: ')
					if nombreH in familia:
						childrenss = list(node_p.children)
						childrenss.append(familia[nombreH])
						node_p.children = tuple(childrenss)
					else:
						familia[nombreH] = Node(str(nombreH))
						childrenss = list(node_p.children)
						childrenss.append(familia[nombreH])
						node_p.children = tuple(childrenss)
			else:
				print('Nuevo Integrante')
				rol_familiar = input('Es un (hijo/padre): ')
				if rol_familiar == 'hijo':
					nombreP = input('De quien: ')
					if nombreP in familia:
						familia[nombre] = Node(str(nombre))
						familia[nombre].parent = familia[nombreP]
					else:
						familia[nombreP] = Node(nombreP)
						familia[nombre] = Node(str(nombre))
						familia[nombre].parent = familia[nombreP]
						
				elif rol_familiar == 'padre':
					nombreH = input('De quien: ')
					if nombreH in familia:
						familia[nombre] = Node(str(nombre))
						familia[nombre].children=[familia[nombreH]]
					else:
						familia[nombre] = Node(str(nombre))
						familia[nombreH] = Node(str(nombreH))
						familia[nombre].children = [familia[nombreH]]
			renderizar_familia()
		elif accion == '2':
			print('Ingrese el Nombre y rol que quiere saber\nEjemplo: \n [Nombre],[Rol]\n')
			termino_buscar = input('Ingrese el Nombre y rol que quiere saber')
			terminos = termino_buscar.split(",")
			persona = terminos[0]
			relacion = terminos[1]
			if persona in familia:
				if relacion == 'hijos':
					buscar_hijos(familia[persona])
				elif relacion == 'padre':
					buscar_padre(familia[persona])
				elif relacion == 'tios':
					buscar_tios(familia[persona])
				elif relacion == 'hermanos':
					buscar_hermanos(familia[persona])
				elif relacion == 'abuelo':
					buscar_abuelo(familia[persona])
				elif relacion == 'nietos':
					buscar_nietos(familia[persona])
			else:
				print(f'{persona} no pertenece a la familia')
		else:
			ejercicio == '1'
	renderizar_familia()
```
